# 🥇 Ranking de ventas por categoría

## 📌 Problema

Dado un conjunto de ventas, asigna un ranking dentro de cada categoría basado en el monto de la venta (de mayor a menor).

## 🧠 Concepto clave

* `RANK()`
* `PARTITION BY`
* `ORDER BY`


## 📊 Query


```sql
SELECT *,
       RANK() OVER (PARTITION BY category ORDER BY amount DESC) AS rank_in_category
FROM sales;
```

## 📈 Explicación

Se divide el dataset por categoría y se ordena por monto de venta en orden descendente.
Luego se asigna un ranking donde valores iguales comparten la misma posición.

## 🎯 Caso de uso real

* Top productos por categoría
* Ranking de vendedores
* Clasificación de desempeño

## ⚠️ Nota

Si hay empates, el ranking se salta números (ej: 1,1,3).
